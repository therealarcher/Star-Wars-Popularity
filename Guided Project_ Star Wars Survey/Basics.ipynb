```python
import numpy as np
movie_mapping = {
    'Star Wars: Episode I  The Phantom Menace': True,
    np.nan: False,
    'Star Wars: Episode II  Attack of the Clones': True,
    'Star Wars: Episode III  Revenge of the Sith': True,
    'Star Wars: Episode IV  A New Hope': True,
    'Star Wars: Episode V The Empire Strikes Back': True,
    'Star Wars: Episode VI Return of the Jedi': True 
}

for col in star_wars.columns[3:9]:
    star_wars[col] = star_wars[col].map(movie_mapping)
```

```python
star_wars = star_wars.rename(columns = {
    "Which of the following Star Wars films have you seen? Please select all that apply.": "seen_1",
    "Unnamed: 4": "seen_2",
    "Unnamed: 5": "seen_3",
    "Unnamed: 6": "seen_4",
    "Unnamed: 7": "seen_5",
    "Unnamed: 8": "seen_6"   
})
```

 # Ranking of movies

```python
star_wars[star_wars.columns[9:15]] =  star_wars[star_wars.columns[9:15]].astype(float)
star_wars = star_wars.rename(columns = {
    "Please rank the Star Wars films in order of preference with 1 being your favorite film in the franchise and 6 being your least favorite film.": "ranking_1",
    "Unnamed: 10": "ranking_2",
    "Unnamed: 11": "ranking_3",
    "Unnamed: 12": "ranking_4",
    "Unnamed: 13": "ranking_5",
    "Unnamed: 14": "ranking_6"
})
star_wars.head()
```

```python
star_wars[star_wars.columns[9:15]].mean()
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.bar([1,2,3,4,5,6], star_wars[star_wars.columns[9:15]].mean(), align="center")
plt.xticks([1,2,3,4,5,6])
plt.xlabel("Episode number")
plt.ylabel("Mean ranking (lower is better)")
```

As suspected, episodes 4-6 of Star Wars are more popular than the the first 3, episode 3 being the least popular, and episode 5 being the most.

# View counts

```python
star_wars[star_wars.columns[3:9]].sum()
```

```python
plt.bar([1,2,3,4,5,6], star_wars[star_wars.columns[3:9]].sum(), align = "center")
plt.xticks([1,2,3,4,5,6])
plt.xlabel("Episode number")
plt.ylabel("Number of views")
plt.title("Number of views of each episode")
```

It appears as though that episode 1, 5, and 6 got the most views in the sample collected.

## Gender differences in popularity

```python
males = star_wars[star_wars["Gender"] == "Male"]
females = star_wars[star_wars["Gender"] == "Female"]
plt.xlabel("Episode number")
plt.ylabel("Mean ranking (lower is better)")
plt.xticks([1,2,3,4,5,6])
plt.title("Male only popularity")
plt.bar([1,2,3,4,5,6], males[star_wars.columns[9:15]].mean(), align="center")
plt.show()
plt.xlabel("Episode number")
plt.ylabel("Mean ranking (lower is better)")
plt.xticks([1,2,3,4,5,6])
plt.title("Female only popularity")
plt.bar([1,2,3,4,5,6], females[star_wars.columns[9:15]].mean(), align="center")
plt.show()




```

The trend is quite similar between males and females.  Females seemed to prefer episode one more than men did.

```python
plt.bar([1,2,3,4,5,6], males[star_wars.columns[3:9]].sum(), align = "center")
plt.xticks([1,2,3,4,5,6])
plt.xlabel("Episode number")
plt.ylabel("Number of views")
plt.title("Male views of each episode")
plt.show()

plt.bar([1,2,3,4,5,6], females[star_wars.columns[3:9]].sum(), align = "center")
plt.xticks([1,2,3,4,5,6])
plt.xlabel("Episode number")
plt.ylabel("Number of views")
plt.title("Female views of each episode")
```

It appears as though males viewed the first 3 episodes more than females did (yet preferred them less).
